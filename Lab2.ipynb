```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import scipy
import math
from scipy import stats
from scipy.stats import poisson
plt.rcParams["figure.figsize"] = (15,10)
```

```python
print('Problem 1')
print('Problem A')
print('The x is 3 and y is 7.5. As i integrate for more days, the peak of the probability distribution of background get lower.')
print('Problem B')
print('It makes sense that after 5 days, the summed probability distribution is still in Poisson distribution. Mathematically, the result of convolution shows that each function was Poisson distribution before, so it must still be Poisson distribution. Conceptually, the probability distribution is always the same, no matter how many days we take.')
average_cos_ray = 3
average_gamma_ray = 7.5
x = np.arange(10)
pmf = stats.poisson.pmf(x,average_cos_ray)
y = pmf
plt.step(x, y,label = 'Day1')
y = np.convolve(y,pmf)
x = np.arange(len(y))
plt.step(x, y,label = 'Day2')
y = np.convolve(y,pmf)
x = np.arange(len(y))
plt.step(x, y,label = 'Day3')
y = np.convolve(y,pmf)
x = np.arange(len(y))
plt.step(x, y, label = 'Day4')
y = np.convolve(y,pmf)
x = np.arange(len(y))
plt.step(x, y,label = 'Day5')
plt.xlabel("Cosmic Ray Background")
plt.ylabel("Probability")
plt.legend()
```
