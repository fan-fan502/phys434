```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import scipy
from scipy import stats
#this sets the size of the plot to something useful
plt.rcParams["figure.figsize"] = (15,10)
```

```python
d = stats.norm.rvs(loc = 5., scale = 0.01, size = 100000)
```

```python
print('Q1')
print(stats.norm.cdf(0,0,1))
print('z = (0-0)/1 = 0 according to the z-table, the percentage is 0.5')
print(stats.norm.cdf(1,0,2))
print('z = (1-0)/2 = 0.5 according to the z-table, the percentage is 0.6915')
print(stats.norm.cdf(4,3,4))
print('z = (4-3)/4 = 0.25 according to the z-table, the percentage is 0.5987')
print(stats.norm.ppf(0.95,0,1))
print('0.95 in the z table is z = 1.65')
print(stats.norm.ppf(0.9207,0,1))
print('0.95 in the z table is z = 1.41')
from scipy.stats import gamma
import matplotlib.pyplot as plt
gamma.pdf(6, 3, 3, 2)
from scipy.stats import gamma
import matplotlib.pyplot as plt
gamma.pdf(6, 5, 3, 2)
```

```python
print('Q2')
d = stats.expon.rvs(loc = 5., scale = 0.01,size = 10000)
fig, ax = plt.subplots(1, 1)
ax.hist(d,50, density=True)
plt.tick_params(labelsize = 24)
plt.xlim([4.95,5.05])
x = np.linspace(4.95,5.05,1000)
ax.plot(x,stats.expon.pdf(x,loc = 5., scale = 0.01),linewidth = 8,alpha = 0.7)
plt.show()
```

```python
print('Q3')
print('i want to know what is the probability of the value 5.02?')
print(1- stats.expon.cdf(5.02,5,4))
print('i want to know what is the sigma?')
print(stats.norm.ppf(0.995,0,1))
```

```python
print('Q4')
print(1- stats.expon.cdf(4.98,5,4))
print(1- stats.expon.cdf(5,5,4))
print(1- stats.expon.cdf(6,5,4))
print(1- stats.expon.cdf(7,5,4))
print('The pattern is that the bigger the x value the smaller the probability is')
```

```python
print('Q5')
r = stats.poisson.rvs(5,size = 1000)
fig, ax = plt.subplots(1, 1)
ax.hist(r,50, density=True)
plt.tick_params(labelsize = 24)
plt.xlim([0,15])
plt.show()
```

```python
print('i want to know what is the probability of the value 6?')
print(1- stats.poisson.cdf(6,5,0))
print('i want to know what is the sigma?')
print(stats.norm.ppf(0.2378,0,1))
print('The reason why mean value can be greater is because the boundary number can be infinite big and that can influence the mean value')
```
